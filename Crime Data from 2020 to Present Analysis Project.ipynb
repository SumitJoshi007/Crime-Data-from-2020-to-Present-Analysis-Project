## About Dataset :- 

This dataset reflects incidents of crime in the City of Los Angeles dating back to 2020. This data is transcribed from original crime reports that are typed on paper and therefore there may be some inaccuracies within the data. Some location fields with missing data are noted as (0°, 0°). Address fields are only provided to the nearest hundred block in order to maintain privacy. This data is as accurate as the data in the database.

## 3. Can you identify the age and gender distribution of victims in the dataset?

```python
def age_group(age):
    if(age>=0 and age<10):
        return '0 to 10'
    elif(age>=10 and age<20):
        return '10 to 20'
    elif(age>=20 and age<30):
        return '20 to 30'
    elif(age>=30 and age<40):
        return '30 to 40'
    elif(age>=40 and age<50):
        return '40 to 50'
    elif(age>=50 and age<60):
        return '50 to 60'
    elif(age>=60 and age<70):
        return '60 to 70'
    elif(age>=70 and age<80):
        return '70 to 80'
    elif(age>=80 and age<90):
        return '80 to 90'
    else:
        return '90 above'

df['age group'] = df['Vict Age'].apply(age_group)
df
```

```python
df['age group'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Vict Age'],rwidth=0.9,color='red')
plt.xlabel('Age Group')
plt.ylabel('No. of Cases')
plt.grid()
plt.show()
```

## 7. what are the trend over the time period covered by dataset ? 

```python
ndf = df['DATE OCC'].value_counts()
ndf = ndf.reset_index()
ndf = ndf.rename(columns={'index': 'DATE OCC', 'DATE OCC': 'count'})
ndf = ndf.sort_values(by='DATE OCC')
ndf
```

```python
y = ndf['count'].max()                                     #highest case denoted as y.
ddf = ndf.loc[ndf['count'] == y]
x = ddf['DATE OCC'].values[0]
```

```python
y1 = ndf['count'].min()                                     #lowest case denoted as y1.
ddf = ndf.loc[ndf['count'] == y1]
x1 = ddf['DATE OCC'].values[0]
```

```python
plt.figure(figsize=(15,5),facecolor='lightgray')
plt.plot(ndf['DATE OCC'],ndf['count'],color='red')
plt.title('Crime Trend')
plt.xlabel("Time")
plt.ylabel('Frequency')
plt.grid()
plt.xticks(rotation=20)
plt.text(x,y,f' {   str(x).split("T")[0]   } {y}')
plt.text(x1,y1,f'{   str(x1).split("T")[0]   } {y1}')
plt.show()
```

```python
plt.figure(figsize=(15,5),facecolor='lightgray')
plt.scatter(ndf['DATE OCC'],ndf['count'],color='red',alpha=0.5)
plt.title('Crime Trend')
plt.xlabel("Time")
plt.ylabel('Frequency')
plt.grid()
plt.xticks(rotation=20)
plt.show()
```

# Over Time Period

## Yearly Trend

```python
df['Year'] = df['DATE OCC'].dt.year
ndf = df['Year'].value_counts()
ndf= ndf.reset_index()
ndf= ndf.rename(columns={'index': 'Year', 'Year': 'count'})
ndf = ndf.sort_values(by='Year')
ndf
```

```python
plt.figure(figsize=(15,5),facecolor='lightgray')
plt.plot(ndf['Year'],ndf['count'],color='firebrick')
plt.title('Yearly Crime Trend')
plt.xlabel("Year")
plt.ylabel('Frequency')
plt.grid()
plt.xticks(rotation=20)
plt.show()
```

## Each Year Month with No. of Crimes

```python
df['Month'] = df['DATE OCC'].dt.month_name()
ndf = df.groupby(['Year', 'Month']).agg(
result = ('DR_NO', 'count')
)
ndf = ndf.reset_index()
ndf = ndf.sort_values(by=['Year','result'], ascending=[True,False])
ndf
```

```python
sns.catplot(data=ndf, x='Year', y='result', kind='bar', hue='Month', aspect=3)
plt.title('Age Group')
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.grid()
plt.xticks(rotation=20)
plt.show()
```
